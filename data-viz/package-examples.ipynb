# Data Visualisation in Python

This notebook demonstrates 3 packages: matplotlib, plotly and altair.

## Import packages

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import altair as alt
```

## Line chart

```python
# Create the Data
x = [1, 2, 3, 4, 5]
y = [10, 15, 7, 12, 9]
```

### matplotlib

```python
# Create the plot
plt.plot(x, y)

# Set labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Chart Example')

# Display the plot
plt.show()
```

### plotly

```python
# Create the plot
fig = go.Figure(data=go.Scatter(x=x, y=y))

# Set labels and title
fig.update_layout(title='Line Chart Example', xaxis_title='X-axis', yaxis_title='Y-axis')

# Display the plot
fig.show()
```

### altair

```python
data = pd.DataFrame({'x': x, 'y': y})

# Create the plot
chart = alt.Chart(data).mark_line().encode(
    x='x',
    y='y'
).properties(
    title='Line Chart Example'
)

# Display the plot
chart
```

## Bar chart

```python
# Data
categories = ['A', 'B', 'C', 'D']
values = [20, 35, 30, 25]
```

### matplotlib

```python
# Create the plot
plt.bar(categories, values)

# Set labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart Example')

# Display the plot
plt.show()
```

### plotly

```python
# Create the plot
fig = go.Figure(data=go.Bar(x=categories, y=values))

# Set labels and title
fig.update_layout(title='Bar Chart Example', xaxis_title='Categories', yaxis_title='Values')

# Display the plot
fig.show()
```

### altair

```python
# Data
data = pd.DataFrame({'categories': categories, 'values': values})

# Create the plot
chart = alt.Chart(data).mark_bar().encode(
    x='categories',
    y='values'
).properties(
    title='Bar Chart Example'
)

# Display the plot
chart
```

## Pie chart

```python
labels = ['A', 'B', 'C', 'D', 'E']
sizes = [5, 10, 20, 30, 35]  # Values for each slice
```

### matplotlib

```python
# Create the plot
plt.pie(sizes, labels=labels)

# Set title
plt.title('Pie Chart Example')

# Display the plot
plt.show()
```

### plotly

```python
# Create the plot
fig = go.Figure(data=go.Pie(labels=labels, values=sizes))

# Set title
fig.update_layout(title='Pie Chart Example')

# Display the plot
fig.show()
```

### altair

```python
# Data
data = pd.DataFrame({'sizes': sizes, 'labels': labels})

# Create the plot
chart = alt.Chart(data).mark_arc().encode(
    theta='sizes',
    color='labels'
).properties(
    title='Pie Chart Example'
)

# Display the plot
chart
```

## Advanced plots

Plotly & Altair are primarily focused on provide interactive and declarative visualisations, limiting the number of plot types available. 

Matplotlib offers low-level control and extensive functionality to create plots such as:

* 3D surface plot with contour lines
* Polar plot
* Streamplot (for visualising fluid flow)
* Treemap (for hierarchical data visualisation)


### 3D surface plot with contour lines

```python
# Generate data for the plot
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Create a 3D surface plot with contour lines
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.contour(X, Y, Z, zdir='z', offset=-1.2, cmap='coolwarm')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Surface Plot with Contour Lines')

# Show the plot
plt.show()
```

### Polar plot

```python
theta = np.linspace(0, 2*np.pi, 100)
radius = np.cos(3*theta)

fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
ax.plot(theta, radius)
ax.set_title('Polar Plot')

plt.show()
```

## Interactive plots

Altair and Plotly are designed for more interactive plots. 

A powerful feature in Altair is the ease and simplicity by which plots can be linked and in this example using brushing. 

The same plot can be generated in Plotly but notice the complex syntax required.

### Altair

```python
# Create scatter plot with interactive linked brushing
brush = alt.selection_interval()

scatter = alt.Chart(cars).mark_circle().encode(
    x='Horsepower',
    y='Miles_per_Gallon',
    color=alt.condition(brush, 'Origin', alt.value('lightgray'))
).add_params(
    brush
)

bar = alt.Chart(cars).mark_bar().encode(
    y='Origin',
    color='Origin',
    x='count(Origin)'
).transform_filter(
    brush
)

chart = scatter & bar

chart
```

### Plotly

```python
from plotly.subplots import make_subplots

# Create scatter plot with interactive linked brushing
fig = make_subplots(rows=1, cols=2, subplot_titles=('Scatter Plot', 'Bar Chart'))

scatter = go.Scatter(
    x=cars['Horsepower'],
    y=cars['Miles_per_Gallon'],
    mode='markers',
    marker=dict(
        color='lightgray',
        size=8,
        opacity=0.7,
        line=dict(width=0)
    ),
    showlegend=False,
    hovertemplate=None
)

fig.add_trace(scatter, row=1, col=1)

bar = go.Bar(
    x=cars['Origin'].value_counts().index,
    y=cars['Origin'].value_counts(),
    marker=dict(
        color=['red', 'green', 'blue']
    ),
    showlegend=False,
    hovertemplate=None
)

fig.add_trace(bar, row=1, col=2)

fig.update_layout(
    title_text='Interactive Linked Brushing',
    hovermode='closest',
    showlegend=False
)

fig.update_layout(height=500, width=900)
fig.update_xaxes(title_text='Horsepower', row=1, col=1)
fig.update_yaxes(title_text='Miles per Gallon', row=1, col=1)
fig.update_xaxes(title_text='Origin', row=1, col=2)
fig.update_yaxes(title_text='Count', row=1, col=2)

fig.show()
```

Plotly does not provide built-in linked brushing functionality between different plot types.

## Interactive 3D visualisations

 Plotly provides powerful tools for creating interactive 3D plots that allow users to explore and interact with data in three dimensions.

```python
# Generate random data
np.random.seed(0)
n_points = 100
x = np.random.rand(n_points)
y = np.random.rand(n_points)
z = np.random.rand(n_points)
color = np.random.rand(n_points)
size = np.random.rand(n_points) * 10

# Create 3D scatter plot
fig = go.Figure(data=go.Scatter3d(
    x=x,
    y=y,
    z=z,
    mode='markers',
    marker=dict(
        color=color,
        size=size,
        opacity=0.8,
        colorscale='Viridis',
        colorbar=dict(title='Color'),
        symbol='circle'
    ),
))

fig.update_layout(
    title='Interactive 3D Scatter Plot',
    scene=dict(
        xaxis_title='X',
        yaxis_title='Y',
        zaxis_title='Z',
    )
)

fig.show()
```

## Animated Plots

While Matplotlib and Altair also have animation capabilities, Plotly provides a more streamlined and interactive approach to creating animations, allowing for more advanced control and customization.

```python
import plotly.graph_objects as go
import numpy as np

# Generate random data
np.random.seed(0)
n_points = 50
x = np.linspace(0, 10, n_points)
y = np.random.randn(n_points)

# Create animated line plot
fig = go.Figure(
    data=go.Scatter(
        x=x,
        y=y,
        mode='lines',
        line=dict(color='blue', width=2)
    ),
    layout=go.Layout(
        title='Animated Line Plot',
        xaxis=dict(range=[0, 10]),
        yaxis=dict(range=[-3, 3]),
        updatemenus=[
            dict(
                type='buttons',
                buttons=[
                    dict(label='Play',
                         method='animate',
                         args=[None, {'frame': {'duration': 500, 'redraw': True}, 'fromcurrent': True, 'transition': {'duration': 0}}]),
                    dict(label='Pause',
                         method='animate',
                         args=[[None], {'frame': {'duration': 0, 'redraw': False}, 'mode': 'immediate', 'transition': {'duration': 0}}])
                ],
                visible=True,
                direction='left',
                x=0.1,
                y=1.1
            )
        ]
    ),
    frames=[
        go.Frame(data=go.Scatter(x=x[:i], y=y[:i]), name=f'frame_{i}') for i in range(1, n_points)
    ]
)

fig.show()
```
